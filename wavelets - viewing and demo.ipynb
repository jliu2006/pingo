```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sio
```

```python
def plot(image, brightness=2):
    newimage = np.transpose(image[0:3, :, :], axes=(1, 2, 0))/10000
    plt.axis("off")
    plt.imshow(brightness*newimage)
    plt.show()
```

```python
wavelet = "wv32"
bandtype = 3
```

```python
def getData(band1, band2):
    return sio.loadmat(f"../data/thermokarst_lakes_preprocessed/wavelet-decomposed-py/{wavelet}/band_{band1}_{band2}.mat")["imgs"]
```

```python
datashape = getData(0, 0).shape; datashape
```

```python
imgs = np.zeros((bandtype, bandtype) + datashape)

for i in range(bandtype):
    for j in range(bandtype):
        imgs[i][j] = getData(i, j)
```

```python
fullDecomp = np.concatenate(np.concatenate(imgs, 3), 3)
```

```python
plot(imgs[0][0][0], 4/bandtype)
```

```python
plot(fullDecomp[0], 4/bandtype)
```

```python
plot(fullDecomp[0], 4/bandtype*64)
```
